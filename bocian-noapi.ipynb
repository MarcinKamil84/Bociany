```python
df_2022 = pd.read_csv('bocian_2022.csv')

df_2022_subset = df_2022.iloc[:, [0, 1, 2, 3, 4, 13]]

df_2022_subset.dropna(how='all', inplace=True)

df_2022_subset.columns = ['lp', 'miejscowosc', 'gmina', 'gps', 'podstawa', 'stan_2022']

df_2022_subset[['lat', 'lon']] = df_2022_subset['gps'].str.split(',', expand=True).astype('float')

df_2022_subset['mlode_2022'] = df_2022_subset.stan_2022.str.extract(r'((?<=HPm)\d)')

df_2022_subset = df_2022_subset[df_2022_subset['stan_2022'] != 'brak']

df_2022_subset = df_2022_subset.fillna(0)
```

## Combining 2021 and 2022

```python
df_full = df_2021_subset.merge(df_2022_subset, how='outer')

df_full.fillna(0, inplace=True)

df_full['mlode_2022'] = df_full['mlode_2022'].astype('int')
df_full['mlode_2021'] = df_full['mlode_2021'].astype('int')

df_full['delta_2021_2022'] = df_full['mlode_2022'] - df_full['mlode_2021']
df_full['suma_2021_2022'] = df_full['mlode_2022'] + df_full['mlode_2021']


conditions = [
    (df_full['mlode_2021'] == 0) & (df_full['mlode_2022'] == 0),
    (df_full['mlode_2021'] != 0) & (df_full['mlode_2022'] == 0),
    (df_full['mlode_2021'] == 0) & (df_full['mlode_2022'] != 0),
    (df_full['mlode_2021'] != 0) & (df_full['mlode_2022'] != 0)
]

choices = [
    'brak młodych',
    'tylko 2021',
    'tylko 2022',
    '2021 i 2022'
]

df_full['lata_z_mlodymi'] = np.select(conditions, choices, "brak danych")

df_full['stan_2022'] = df_full['stan_2022'].astype('str')
df_full['stan_2021'] = df_full['stan_2021'].astype('str')

df_full = df_full[(df_full.stan_2021 != '0') & (df_full.stan_2022 != '0')]
```

```python
lat = (df_full['lat'].astype('float').min() + df_full['lat'].astype('float').max()) / 2
lon = (df_full['lon'].astype('float').min() + df_full['lon'].astype('float').max()) / 2
```

```python
def grouping(df, vals, gp):
    '''
    Groups dataframe by given columns
    
    Parameters:
    df: dataframe to apply to
    values: columns to  compare
    gp: column to group by
    
    Returns:
    grouped dataframe
    '''
    
    new_vals = vals[:]
    new_vals.append(gp)
    
    df_gpd = df[new_vals].groupby(gp).sum().stack().unstack()
    
    df_gpd.index = df_gpd.index.astype('str')
    
    return df_gpd
```

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.transform import dodge
from bokeh.palettes import Paired

def compare_graph(df, gr, values, ylabel, ptitle):
    
    '''
    Draws graph to compare years
    
    Parameters:
    df: original dataframe
    gr: column to group
    values (list): values to compare
    ylabel: y-axis label
    '''

    new_df = grouping(df_full, values, gr)
    
    source = ColumnDataSource(data=new_df)

    p = figure(x_range=list(new_df.index), y_range=(0, new_df[values].to_numpy().max()*1.1), title=ptitle,
               width=900, toolbar_location=None, tools='hover', tooltips='@$name')
    
    p.vbar(x=dodge(gr, -0.12, range=p.x_range), top='mlode_2021', source=source, name="mlode_2021",
           width=0.2, color=Paired[3][0], legend_label="2021")

    p.vbar(x=dodge(gr, 0.12, range=p.x_range), top='mlode_2022', source=source, name="mlode_2022",
           width=0.2, color=Paired[3][1], legend_label="2022")
    
    p.yaxis.axis_label = ylabel
    
    p.xaxis.major_label_orientation = 0.3
    
    p.legend.location = "top_left"
    p.legend.orientation = "horizontal"
    
    p.background_fill_color = Paired[4][0]
    p.background_fill_alpha = 0.1

    reset_output()
    output_notebook()
    
    show(p)
    
    output_file(filename="{}.html".format(ptitle), title="Bociany: suma liczby młodych w latach 2021-2022")

    save(p)
```

# Plotting maps

```python
from bokeh.io import show, reset_output
from bokeh.transform import linear_cmap
from bokeh.palettes import PuRd as palette
from bokeh.plotting import gmap, save, output_file
from bokeh.models import ColorBar
from bokeh.models import GMapOptions, ColumnDataSource, HoverTool
```

```python
def plot_v1(lat, lng, zoom=11, map_type='satellite'):
    
    # Google Maps options
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    
    # Behaviour on hover
    hover = HoverTool(
        tooltips = [
            ('Numer', '@lp'),
            ('Podstawa', '@podstawa'),
            ('Młode 2021', '@mlode_2021'),
            ('Młode 2022', '@mlode_2022'), 
        ]
    )

    p = gmap(api_key, gmap_options, title='Bociany: suma liczby młodych w latach 2021-2022',
        width=900, height=600,
        tools=[hover, 'reset', 'wheel_zoom', 'pan'])

    # Data source
    source = ColumnDataSource(df_full)
    
    # Colour mapping
    mapper = linear_cmap('suma_2021_2022', palette[9][::-1], df_full['suma_2021_2022'].min(), df_full['suma_2021_2022'].max())
    
    # Data points
    center = p.circle('lon', 'lat', size=8, alpha=0.7, 
                      color=mapper, source=source,
                      line_color='#000000'
                     )
    
    color_bar = ColorBar(color_mapper=mapper['transform'], location=(0,0))
    p.add_layout(color_bar, 'right')

    reset_output()
    output_notebook()

    show(p)
    
    output_file(filename="bociany01.html", title="Bociany: suma liczby młodych w latach 2021-2022")

    save(p)

    return p

p = plot_v1(lat, lon, map_type='terrain')
```

```python
from bokeh.palettes import Category10_4 as palette

df_full = df_full.sort_values('lata_z_mlodymi')

def plot_v3(lat, lng, zoom=11, map_type='roadmap'):
    
    # Google Maps options
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    
    # Behaviour on hover
    hover = HoverTool(
        tooltips = [
            ('Numer', '@lp'),
            ('Podstawa', '@podstawa'),
            ('Młode 2021', '@mlode_2021'),
            ('Młode 2022', '@mlode_2022'), 
        ]
    )

    p = gmap(api_key, gmap_options, title='Bociany: młode na gniazdach wg. lat',
        width=900, height=600,
        tools=[hover, 'reset', 'wheel_zoom', 'pan'])

    # Data source
    sources = [
        ColumnDataSource(df_full[df_full['lata_z_mlodymi']==df_full['lata_z_mlodymi'].unique()[0]]),
        ColumnDataSource(df_full[df_full['lata_z_mlodymi']==df_full['lata_z_mlodymi'].unique()[1]]),
        ColumnDataSource(df_full[df_full['lata_z_mlodymi']==df_full['lata_z_mlodymi'].unique()[2]]),
        ColumnDataSource(df_full[df_full['lata_z_mlodymi']==df_full['lata_z_mlodymi'].unique()[3]]),
        ]
    
    # Colour mapping
    mapper = linear_cmap('lata_z_mlodymi', palette, 0, 3)
    
    
    # Data points
    n = [0, 1, 2, 3]
    for n, i in zip(n, sources):
        p.circle('lon', 'lat', size=8, alpha=0.7,
                 color=palette[n], source=i,
                 legend_group='lata_z_mlodymi',
                 line_color='#000000'
                )
        
    p.legend.click_policy="mute"
    p.legend.title = 'Lata z młodymi'
    p.legend.location = "top_left"

    reset_output()
    output_notebook()
    
    show(p)
    
    output_file(filename="bociany03.html", title="Bociany: młode na gniazdach wg. lat")
    
    save(p)
    
    return p

p = plot_v3(lat, lon, map_type='satellite')
```
